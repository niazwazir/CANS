```python
######################### AUTOMATICALLY CREATED PATCHES OF DIFFERENT SR METHODS SUCH AS SRCNN, VDSR, FSRCNN, MDSR, PAN, CARN, A2N METHODS
import numpy as np
import cv2
import os

def create_patches(image_folder, output_folder, patch_size=(100,100)):
    """Create patches from images in a specified folder and save them to an output folder."""
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    patch_index = 1
    for filename in os.listdir(image_folder):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            img = cv2.imread(os.path.join(image_folder, filename))
            if img is None:
                continue  # Skip if the image is not read properly
            h, w, _ = img.shape
            for i in range(0, h - patch_size[0] + 1, patch_size[0]):
                for j in range(0, w - patch_size[1] + 1, patch_size[1]):
                    patch = img[i:i + patch_size[0], j:j + patch_size[1]]
                    patch_filename = f'patch_{patch_index}.png'
                    cv2.imwrite(os.path.join(output_folder, patch_filename), patch)
                    patch_index += 1

def main():
    # Define the folders containing the images
    folder1 = 'HR_SCALE_X3'  # Path to the first folder of images
    folder2 = 'MDSR_METHOD_X3'  # Path to the second folder of images
    folder3 = 'PAN_METHOD_X3'  # Path to the second folder of images
    folder4 = 'CARN_METHOD_X3'  # Path to the second folder of images
    folder5 = 'A2N_METHOD_X3'  # Path to the second folder of images

    # Create patches from both folders and save them to new folders
    create_patches(folder1, output_folder='1_PATCH_HR_SCALE_X3')
    create_patches(folder2, output_folder='2_PATCH_MDSR_METHOD_X3')
    create_patches(folder3, output_folder='3_PATCH_PAN_METHOD_X3')
    create_patches(folder4, output_folder='4_PATCH_CARN_METHOD_X3')
    create_patches(folder5, output_folder='5_PATCH_A2N_METHOD_X3')

if __name__ == '__main__':
    main()
```

```python
#CALCULATE PSNR OF EACH PATCH AND AVERAGE OF EACH METHOD PSNR
import numpy as np
import cv2
import os

def calculate_psnr(original, reconstructed):
    """Calculate PSNR between two images."""
    mse = np.mean((original - reconstructed) ** 2)
    if mse == 0:
        return 100  # Infinite PSNR
    return 20 * np.log10(255.0 / np.sqrt(mse))

def compute_average_psnr(folder1, folder2):
    """Compute PSNR and average PSNR between two folders."""
    images1 = sorted(os.listdir(folder1))
    images2 = sorted(os.listdir(folder2))

    # Ensure both folders have the same number of images
    if len(images1) != len(images2):
        print(f"The number of images in {folder1} and {folder2} must be the same.")
        return None

    total_psnr = 0
    num_images = len(images1)

    # Calculate PSNR for each pair of images
    for img1, img2 in zip(images1, images2):
        # Read images
        image1 = cv2.imread(os.path.join(folder1, img1))
        image2 = cv2.imread(os.path.join(folder2, img2))

        # Check if images are loaded properly
        if image1 is None or image2 is None:
            print(f"Error loading images: {img1} or {img2}. Skipping.")
            continue

        # Calculate PSNR
        psnr_value = calculate_psnr(image1, image2)
        total_psnr += psnr_value

        # Print PSNR for the current image pair
        print(f'PSNR for {img1} vs {img2}: {psnr_value:.2f} dB')

    # Calculate average PSNR
    average_psnr = total_psnr / num_images
    return average_psnr

def main():
    # Define the folders containing the images
    folder1 = '1_PATCH_HR_SCALE_X3'  # Replace with your first folder path
    folder2 = '2_PATCH_MDSR_METHOD_X3'  # Replace with your second folder path
    folder3 = '3_PATCH_PAN_METHOD_X3'  # Replace with your third folder path
    folder4 = '4_PATCH_CARN_METHOD_X3'  # Replace with your fourth folder path
    folder5 = '5_PATCH_A2N_METHOD_X3'  # Replace with your fourth folder path

    # Compute and print average PSNR for each comparison
    avg_psnr_2 = compute_average_psnr(folder1, folder2)
    if avg_psnr_2 is not None:
        print(f'Average PSNR between {folder1} and {folder2}: {avg_psnr_2:.2f} dB')

    avg_psnr_3 = compute_average_psnr(folder1, folder3)
    if avg_psnr_3 is not None:
        print(f'Average PSNR between {folder1} and {folder3}: {avg_psnr_3:.2f} dB')

    avg_psnr_4 = compute_average_psnr(folder1, folder4)
    if avg_psnr_4 is not None:
        print(f'Average PSNR between {folder1} and {folder4}: {avg_psnr_4:.2f} dB')
    
    avg_psnr_5 = compute_average_psnr(folder1, folder5)
    if avg_psnr_5 is not None:
        print(f'Average PSNR between {folder1} and {folder5}: {avg_psnr_5:.2f} dB')

if __name__ == '__main__':
    main()
```

```python
import numpy as np
import cv2
import os
from skimage.metrics import structural_similarity as ssim

def compute_average_ssim(hr_folder, compare_folder):
    """Compute SSIM and average SSIM between HR folder and another folder."""
    hr_images = sorted(os.listdir(hr_folder))
    compare_images = sorted(os.listdir(compare_folder))

    # Ensure both folders have the same number of images
    if len(hr_images) != len(compare_images):
        print(f"The number of images in {hr_folder} and {compare_folder} must be the same.")
        return None

    total_ssim = 0
    num_images = len(hr_images)

    # Calculate SSIM for each pair of images
    for hr_img, compare_img in zip(hr_images, compare_images):
        # Read images
        hr_image = cv2.imread(os.path.join(hr_folder, hr_img))
        compare_image = cv2.imread(os.path.join(compare_folder, compare_img))

        # Check if images are loaded properly
        if hr_image is None or compare_image is None:
            print(f"Error loading images: {hr_img} or {compare_img}. Skipping.")
            continue

        # Convert images to grayscale for SSIM calculation
        hr_image_gray = cv2.cvtColor(hr_image, cv2.COLOR_BGR2GRAY)
        compare_image_gray = cv2.cvtColor(compare_image, cv2.COLOR_BGR2GRAY)

        # Calculate SSIM
        ssim_value = ssim(hr_image_gray, compare_image_gray)
        total_ssim += ssim_value

        # Print SSIM for the current image pair
        print(f'SSIM for {hr_img} vs {compare_img}: {ssim_value:.4f}')

    # Calculate average SSIM
    average_ssim = total_ssim / num_images
    return average_ssim

def main():
    # Define the folders containing the images
  
    folder1 = '1_PATCH_HR_SCALE_X3'  # Replace with your first folder path
    folder2 = '2_PATCH_MDSR_METHOD_X3'  # Replace with your second folder path
    folder3 = '3_PATCH_PAN_METHOD_X3'  # Replace with your third folder path
    folder4 = '4_PATCH_CARN_METHOD_X3'  # Replace with your fourth folder path
    folder5 = '5_PATCH_A2N_METHOD_X3'  # Replace with your fourth folder path

    # Compute and print average SSIM for each comparison
    avg_ssim_wazir1 = compute_average_ssim(folder1, folder2)
    if avg_ssim_wazir1 is not None:
        print(f'Average SSIM between {folder1} and {folder2}: {avg_ssim_wazir1:.4f}')

    avg_ssim_wazir2 = compute_average_ssim(folder1, folder3)
    if avg_ssim_wazir2 is not None:
        print(f'Average SSIM between {folder1} and {folder3}: {avg_ssim_wazir2:.4f}')

    avg_ssim_wazir3 = compute_average_ssim(folder1, folder4)
    if avg_ssim_wazir3 is not None:
        print(f'Average SSIM between {folder1} and {folder4}: {avg_ssim_wazir3:.4f}')

    avg_ssim_wazir4 = compute_average_ssim(folder1, folder5)
    if avg_ssim_wazir4 is not None:
        print(f'Average SSIM between {folder1} and {folder5}: {avg_ssim_wazir4:.4f}')

if __name__ == '__main__':
    main()
```
